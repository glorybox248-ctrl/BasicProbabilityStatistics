```
import numpy as np
import matplotlib.pyplot as plt
x = np.array([1,2,3,4,5,6,7])
y = np.array([73.2,74.5,75.8,72.1,71.1,68.2,67.7])
plt.plot(x,y,color="b",marker='o')
plt.grid()
plt.show()
```

```
import numpy as np
import matplotlib.pyplot as plt
x = np.linspace(-5,5,300)
y1= x**2
y2= (x-2)**2
plt.plot(x,y1,color="b")
plt.plot(x,y2,color="k",linestyle="--")

plt.grid()
plt.show()
```
